```python
import pandas as pd

future = pd.read_csv("future_calendar_rates.csv")
past = pd.read_csv("past_calendar_rates.csv")

df = pd.concat([past, future], ignore_index=True)
```

```python
df['date'] = pd.to_datetime(df['date'])
```

```python
df['month'] = df['date'].dt.month
df['year'] = df['date'].dt.year
```

```python
def demand_period(month):
    if month in [6, 7, 8]:          # Summer peak
        return "peak"
    elif month in [12]:             # Holiday peak
        return "peak"
    elif month in [4, 5, 9, 10]:    # Shoulder seasons
        return "shoulder"
    else:                           # Jan, Feb, Mar, Nov
        return "off-peak"

df['demand_period'] = df['month'].apply(demand_period)
```

```python
df.groupby('demand_period')['occupancy'].mean()
```

```python
df.groupby('demand_period')['revenue'].mean()
```

```python
df.groupby('demand_period')['occupancy'].mean()
```

```python
df.groupby('demand_period')['rate_avg'].mean()
```

```python
df.to_csv("calendar_rates.csv", index=False)
```
